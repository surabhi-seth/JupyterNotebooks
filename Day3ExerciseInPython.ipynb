```python
colsforNaNcheck = ['per_free_lunch', 'per_reduced_lunch', 'mean_ela_score', 'mean_math_score']
```

```python
schools = schools.dropna(subset=colsforNaNcheck, how='all')
```

```python
schools[schools.total_enroll.isna()] # I decided to remove this record too
```

```python
schools = schools.dropna(subset=['total_enroll'])
```

```python
cols = ['mean_ela_score', 'mean_math_score'] # Now I am left with NAs in mean_ela_score & mean_math_score
```

```python
schools[cols] = schools.groupby(["county_name"])[cols].transform(lambda x: x.fillna(x.mean())) #I will default to the means of the county
```

```python
schools.isna().sum() #Yay!!
```

```python
schools['std_ela'] = schools.groupby('year')['mean_ela_score'].transform(lambda x: (x - x.mean()) / x.std()) # Standardizing the ela score
```

```python
schools['std_math'] = schools.groupby('year')['mean_math_score'].transform(lambda x: (x - x.mean()) / x.std()) # Standardizing the math score
```
